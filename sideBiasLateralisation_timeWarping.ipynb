```python
# Load events data
import pickle
import numpy as np
import pandas as pd
import matlab.engine
import os

# Load events from CSV
events_df = pd.read_csv('C:/Users/Lak Lab/Documents/Github/sideBiasLateralisation/analysis/2025-06-13_MBL014_3/2025-06-13_3_MBL014_CorrectedeventTimes.csv')
stim_onsets = events_df['stimulusOnsetTime'].values
choices = events_df['choiceStartTime'].values
rewards = events_df['rewardTime'].values
```

```python
# convert to matlab format
eng.addpath('C:/ruta/a/LakLabAnalysis/easy-plot-neural', nargout=0)  # Ajusta la ruta

# Pass the event vectors to MATLAB workspace
eng.workspace['stimulus_onset'] = matlab.double(stim_onsets.reshape(-1, 1).tolist())
eng.workspace['choices'] = matlab.double(choices.reshape(-1, 1).tolist())
eng.workspace['rewards'] = matlab.double(rewards.reshape(-1, 1).tolist())

# Now create the cell array in MATLAB
eng.eval("events = { 'stimulus_onset', stimulus_onset; 'choice', choices; 'reward', rewards; };", nargout=0)

# Pass timestamps (vector column)
t_col = timestamps.reshape(-1, 1)
eng.workspace['t_col'] = matlab.double(t_col.tolist())
```

```python
# Time warping
warped_matrix = []

for i in range(flu.shape[0]):
    x = flu[i, :].reshape(-1, 1)  # column vector
    eng.workspace['x'] = matlab.double(x.tolist())
    # Call the function using eval to avoid cell array issues
    x_warped = eng.eval("easy.EventAlignedAverageTimeWarped_noPlot(x, t_col, events)", nargout=1)
    # Convert the result to numpy and save
    warped_matrix.append(np.array(x_warped))

warped_matrix = np.stack(warped_matrix)  # [n_neurons x warped_time]
```

```python
# save warped matrix
from scipy.io import savemat
savemat('warped_dff_all_neurons.mat', {'warped_matrix': warped_matrix})
```
