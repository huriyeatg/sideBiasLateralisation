```python
#1> load the environment + define the path
%reload_ext autoreload
%autoreload 2

#make data file
print( "Takes <30s. However, it might take >1 mins or so if network is busy...")
import glob
import os
import pandas as pd
from datetime import datetime
import numpy as np
import main_funcs as mfun
import utils_funcs as utils # utils is from Vape - catcher file: 
import matplotlib.pyplot as plt
import pickle



# Get the list of recordings
info = mfun.analysis()
# display the detected session
print( "Total Session: " +  str(info.recordingList .shape[0]))
#info.recordingList.head()

# set matlab API
import matlab.engine
eng = matlab.engine.start_matlab()
print('Matlab engine is set correctly.')

info.recordingList
```

```python
# Filter recordings
cutoff_date = '2025-04-09'

info.recordingList = info.recordingList[info.recordingList['recordingDate'] > cutoff_date]
# Reset index
info.recordingList = info.recordingList.reset_index(drop=True)

# Display filtered sessions
print("Total Session: " + str(info.recordingList.shape[0]))
print("\nFiltered sessions after April 9th, 2025:")
print(info.recordingList[['recordingDate', 'animalID']])
```

```python
# Extract beh data from all sessions into one csv file
# Create empty list to store all dataframes
all_sessions_data = []

# Get animal ID from the first session (assuming all sessions are from the same animal)
animal_id = info.recordingList.animalID.iloc[0]

# Process each session
for ind, recordingDate in enumerate(info.recordingList.recordingDate):
    try:
        # Check if Timeline.mat exists to determine profile
        filenameTimeline = [f for f in glob.glob(info.recordingList.filepathname[ind] + '\\' + info.recordingList.sessionName[ind] + '_Timeline.mat')]
        if len(filenameTimeline) > 0:
            sessionProfile = 'Grating2AFC'
        else:
            sessionProfile = 'Grating2AFC_noTimeline'
        
        # Get behaviour trial data from Block.mat file
        print('Extracting time events--> Profile: ' + sessionProfile + '  Session: ' + info.recordingList.sessionName[ind])
        
        # Call getBehavData with the session name and profile
        data = eng.getBehavData(info.recordingList.sessionName[ind], sessionProfile)
        
        # Convert MATLAB table to pandas DataFrame using writetable
        temp_csv = os.path.join(info.analysisPath, 'temp_data.csv')
        eng.writetable(data, temp_csv, nargout=0)
        df = pd.read_csv(temp_csv)
        
        # Add session information
        df['expRef'] = info.recordingList.sessionName[ind]
        df['sessionNum'] = ind + 1  # Session number based on chronological order
        
        # Append to list
        all_sessions_data.append(df)
        print(f"Successfully processed session {info.recordingList.sessionName[ind]}")
        
        # Remove temporary file
        os.remove(temp_csv)
        
    except Exception as e:
        print(f"Error processing session {info.recordingList.sessionName[ind]}: {str(e)}")
        continue

# Combine all dataframes
if all_sessions_data:
    combined_data = pd.concat(all_sessions_data, ignore_index=True)
    
    # Save to CSV with animal ID in filename
    output_path = os.path.join(info.analysisPath, f'combined_beh_sessions_{animal_id}.csv')
    combined_data.to_csv(output_path, index=False)
    print(f"\nCombined data saved to: {output_path}")
    print(f"Total trials: {len(combined_data)}")
    print(f"Total sessions: {len(all_sessions_data)}")
    
    # Display first few rows
    print("\nFirst few rows of combined data:")
    print(combined_data.head())
else:
    print("No data was extracted from any session.")
```
